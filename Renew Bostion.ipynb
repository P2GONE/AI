```python
from sklearn.model_selection import cross_val_score

# features, target ë°ì´í„° ì •ì˜
y_target = df1['PRICE']
x_data = df1.drop(['PRICE'], axis=1)

# ì„ í˜•íšŒê·€ ê°ì²´ ìƒì„±
lr = LinearRegression()
lr
```

```python
# 5 folds ì˜ ê°œë³„ Negative MSE scores (ìŒìˆ˜ë¡œ ë§Œë“¤ì–´ ìž‘ì€ ì˜¤ë¥˜ ê°’ì´ ë” í° ìˆ«ìžë¡œ ì¸ì‹ë¨)
neg_mse_scores = cross_val_score(lr, x_data, y_target, scoring="neg_mean_squared_error", cv = 5)
    # cvëŠ” êµì°¨ê²€ì¦ì˜ í´ë“œ ìˆ˜
neg_mse_scores
```

```python
# RMSEë¥¼ êµ¬í•˜ê¸° ìœ„í•´ì„  MSE ê°’ì— -1ì„ ê³±í•œ í›„ í‰ê· ì„ ë‚´ë©´ ëœë‹¤
rmse_scores =  np.sqrt(-1*neg_mse_scores)
rmse_scores
# cross_val_score() # shift + tab, tab ìœ¼ë¡œ í•¨ìˆ˜ í™•ì¸
```

```python
# 5 fold ì˜ í‰ê·  RMSE
avg_rmse = np.mean(rmse_scores)
avg_rmse
```

```python
# cross_val_score(scoring="neg_mean_squared_error")ë¡œ ë°˜í™˜ëœ ê°’ì€ ëª¨ë‘ ìŒìˆ˜
print(' 5 folds ì˜ ê°œë³„ Negative MSE scores: ', np.round(neg_mse_scores, 2))
print(' 5 folds ì˜ ê°œë³„ RMSE scores : ', np.round(rmse_scores, 2))
print(f' 5 folds ì˜ í‰ê·  RMSE : {avg_rmse:.3f}')
```

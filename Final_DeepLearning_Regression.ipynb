```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
 
data= pd.read_csv('C:/Users/charl/Downloads/Tetuan City power consumption.csv', index_col=0)
data.head()
```

```python
df=data.drop('Zone 3  Power Consumption',axis=1)
```

```python
df
```

```python
df1=df.drop('Zone 2  Power Consumption',axis=1)
```

```python
df1
```

```python
df1.columns
```

```python
from sklearn.model_selection import train_test_split

X = df1.drop('Zone 1 Power Consumption',axis=1)
y = df1['Zone 1 Power Consumption']

X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, shuffle=True, random_state=12)
```

```python
import numpy as np

mean = np.mean(X_train, axis = 0)

std = np.std(X_train, axis = 0)

X_train = (X_train - mean) / std
X_test = (X_test - mean) / std
```

```python
print(np.any(np.isnan(X_train)))
print(np.any(np.isnan(X_test)))
print(np.any(np.isinf(X_train)))
print(np.any(np.isinf(X_test)))
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout

model = Sequential()
model.add(Dense(40, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(20, activation='relu'))
model.add(Dense(1))  # 출력 노드를 1개로 변경

model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])

model.summary()
```

```python
N_EPOCHS = 100

history = model.fit(X_train, y_train, epochs = N_EPOCHS, batch_size=64, validation_split=0.2)
```

```python
model.fit(X_train, y_train, epochs=100, batch_size=120, verbose=0)
test_mse_score, test_mae_score = model.evaluate(X_test, y_test)
```

```python
plt.figure(figsize=(10,7))
plt.plot(range(N_EPOCHS), history.history['loss'], label='train loss')
plt.plot(range(N_EPOCHS), history.history['val_loss'], label='test loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.plot(range(N_EPOCHS), history.history['mae'], label='mae')
plt.plot(range(N_EPOCHS), history.history['val_mae'], label='test mae')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
